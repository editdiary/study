#**[Lv.3] Spiral Array**

```python
# My solution
n,m = map(int,input().split())
print()

matrix = [[-1 for _ in range(m)] for _ in range(n)]

x, y = 0, 0
dx, dy = 0, 1
count = 0

while matrix[x][y] == -1 :
    matrix[x][y] = count
    count += 1
    x, y = x+dx, y+dy
    if x == n or y == m or matrix[x][y] != -1 :
        x, y = x-dx, y-dy
        dx, dy = dy, -dx
        x, y = x+dx, y+dy

for i in matrix : 
    for j in i :
        print(f'{j:2d}', end=' ')
    print()  
```

```python
# My solution using Numpy
import numpy as np

n,m = map(int,input().split())
print()

matrix = np.full((n,m),-1)

x, y = 0, 0
dx, dy = 0, 1
count = 0

while matrix[x,y] == -1 :
    matrix[x,y] = count
    count += 1
    x, y = x+dx, y+dy
    if x == n or y == m or matrix[x,y] != -1 :
        x, y = x-dx, y-dy
        dx, dy = dy, -dx
        x, y = x+dx, y+dy

for i in matrix :
    for j in i :
        print(f'{j:2d}', end=' ')
    print()
```
