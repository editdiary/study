#**[Lv.1] Multiples of 3 and 5**

```python
# My solution
num_3 = {x for x in range(3,1000,3)}
num_5 = {y for y in range(5,1000,5)}

print(sum(num_3|num_5))
```

#**[Lv.1] 1~1000에서 각 숫자의 개수 구하기**

```python
# My solution
num_dict = {}

for i in range(1,1000+1) :
    for j in str(i) :
        if j in num_dict :
            num_dict[j] += 1
        else :
            num_dict[j] = 1

print(num_dict)
```

```python
# Best solution
count={ x:0 for x in range(0,10) }

for x in range(1,1001):
    for i in str(x):
        count[int(i)]+=1

print(count)
```

#**[Lv.1] 완전수 구하기**

```python
# My solution
def divisor_sub_N(x) :
    divisor = []
    for i in range(1,x) :
        if x % i == 0 :
            divisor.append(i)
    return divisor

N = int(input())

result = []

for i in range(1,N+1) :
    if sum(divisor_sub_N(i)) == i :
        result.append(i)

print(result)
```

```python
# Best solution
num= int(input("숫자를 입력하시오 : "))
print([x for x in range(1, num+1) if x==sum(y for y in range(1, x) if x%y==0)])
```

#**[Lv.1] 게시판 페이징**

```python
# My solution
def paging(x,y) :
    if x == 0 :
        return 0
    elif x % y == 0 :
        return x // y
    else :
        return x // y + 1

while True :
    print('게시물의 총 갯수와 한페이지에 보여줄 게시물의 수를 띄어서 입력해주세요.')
    m,n = map(int,input().split())
    if n >= 1 :
        print(paging(m,n))
        break
    else :
        print('''*** n의 값을 정확히 입력해주세요.***\nn은 1보다 크거나 같은 정수입니다.''')
        print()
        continue
```

```python
# My solution that even considered exceptions using try-except

def paging(x,y) :
    if x == 0 :
        return 0
    elif x % y == 0 :
        return x // y
    else :
        return x // y + 1

while True :
    print('게시물의 총 갯수와 한페이지에 보여줄 게시물의 수를 띄어서 입력해주세요.')
    try :
        m,n = map(int,input().split())
        if n >= 1 :
            print(paging(m,n))
            break
        else :
            print('''*** n의 값을 정확히 입력해주세요.***\nn은 1보다 크거나 같은 정수입니다.''')
            print()
            continue
    except :
        print('*** 값을 올바르게 입력해주세요 ***')
        print()
```

# **[Lv.1] Special Sort**

```python
# My solution
n = list(map(int,input().split()))

nag_num = [x for x in n if x < 0]
pos_num = [y for y in n if y >= 0]

print(nag_num + pos_num)
```

```python
# Best solution
def do(alist):
    return [x for x in alist if x < 0] + [x for x in alist if x >= 0]

print do([-1, 1, 3, -2, 2])
```
