#**[Lv.2] 넥슨 입사문제 중에서**

```python
# My solution

generator_list = []

for i in range(1, 5000) :
    add_num = []
    for j in str(i) :
        add_num.append(j)
    add_num.append(i)
    generator_list.append(sum(map(int,add_num)))

self_num = {x for x in range(1,5000)} - set(generator_list)

print(sum(self_num))
```

```python
# Best solution
sum(set(range(1, 5000)) - {x + sum([int(a) for a in str(x)]) for x in range(1, 5000)})
```

#**[Lv.2] 구글 입사문제 중에서**

```python
# My solution
count = 0

for i in range(1, 10000+1) :
    for j in str(i) :
        if j == '8' :
            count += 1

print(count)
```

```python
# Best solution(short coding)
print(str(list(range(1,10000+1))).count('8'))
```

#**[Lv.2] 문자열 압축하기**

```python
# My solution
word = input()

count = 1
result = []

for i in range(len(word)) :
    if i == len(word)-1 :
        result.extend([word[i],count])
    else :
        if word[i] == word[i+1] :
            count += 1
        else :
            result.extend([word[i],count])
            count = 1

print(''.join(map(str,result)))
```

#**[Lv.2] 다음 입사문제 중에서**

```python
# My solution
n = list(map(int,input('입력하실 숫자를 띄어쓰기로 입력해주세요 >>> ').split()))

for i in range(len(n)-1) :
    sub_m = n[i+1] - n[i]
    if i == 0 :
        sub = sub_m
        result = [(n[i],n[i+1])]
    elif sub_m < sub :
        sub = sub_m
        result = [(n[i],n[i+1])]
    elif sub == sub_m :
        result.append((n[i],n[i+1]))

for k in result :
    print(k, end=' ')
```
