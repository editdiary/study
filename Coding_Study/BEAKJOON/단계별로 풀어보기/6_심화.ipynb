# [2444번] 별 찍기 - 7

```python
N = int(input())

for i in range(1, 2*N) :
    if i > N :
        i = 2*N - i

    print(' '*(N-i), '*'*(2*i-1), sep='')
```

# [1157번] 단어 공부

```python
word = input().upper()
alpha = list(set(word))

count_list = [word.count(x) for x in alpha]

max_alpha_index = [i for i, v in enumerate(count_list) if v == max(count_list)]

if len(max_alpha_index) > 1 :
    print('?')
else :
    print(alpha[max_alpha_index[-1]])
```

# [2941번] 크로아티아 알파벳

```python
word = input()

croatia = ['c=', 'c-', 'dz=', 'd-', 'lj', 'nj', 's=', 'z=']

croatia_count = sum([word.count(i) for i in croatia if i in word])

print(len(word)-croatia_count)
```

```python
# replace() 함수 활용

word = input()

croatia = ['c=', 'c-', 'dz=', 'd-', 'lj', 'nj', 's=', 'z=']

for i in croatia :
    if i in word :
        word = word.replace(i, '0')

print(len(word))
```

# [1316번] 그룹 단어 체커

```python
N = int(input())

group_word = 0

for _ in range(N) :
    word = input()

    alpha = []
    key = True

    for i in word :
        if i not in alpha :
            alpha.append(i)
        elif i != alpha[-1] :
            key = False
            break

    if key :
        group_word += 1

print(group_word)
```

```python
# Referenced Code
ans = 0

for _ in range(int(input())):
    tmp = ''
    for i in input() :
        if i not in tmp :
            tmp+=i
        elif i != tmp[-1] :
            break
    else :
        ans += 1

print(ans)
```

# [25206번] 너의 평점은

```python
total = [input().split() for _ in range(20)]

remove_P = [x for x in total if x[2] != 'P']

credit = [float(x[1]) for x in remove_P]
grade = [x[2] for x in remove_P]

grade_dic = {'A+':4.5, 'A0':4.0, 'B+':3.5, 'B0':3.0, 'C+':2.5, 'C0':2.0, 'D+':1.5, 'D0':1.0, 'F':0.0}

grade_score = [grade_dic[x] for x in grade]

multiple_score = [credit[x] * grade_score[x] for x in range(len(remove_P))]

print(sum(multiple_score) / sum(credit))
```
