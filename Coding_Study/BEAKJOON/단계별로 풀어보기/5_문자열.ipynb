# [11654번] 아스키 코드

```python
number = [str(x) for x in range(10)]
small_alpha = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n',
               'o','p','q','r','s','t','u','v','w','x','y','z']
cap_alpha = [x.upper() for x in small_alpha]

total_list = number + cap_alpha + small_alpha

ascii_result = [x for x in range(48, 58)] + [x for x in range(65,91)] + [x for x in range(97,123)]


ascii_dic = dict(zip(total_list, ascii_result))

data = input()
print(ascii_dic[data])
```

# [10809번] 알파벳 찾기

```python
S = input()

small_alpha = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n',
               'o','p','q','r','s','t','u','v','w','x','y','z']
number = [S.index(x) if x in S else -1 for x in small_alpha ]
print(*number)
```

# [2675번] 문자열 반복

```python
T = int(input())

for _ in range(T) :
    R, S = input().split()
    P = [x*int(R) for x in S]
    print(*P, sep='')
```

# [5622번] 다이얼

```python
alpha_list = ['ABC', 'DEF', 'GHI', 'JKL', 'MNO', 'PQRS', 'TUV', 'WXYZ']
num_list = []

word = input()
for i in word :
    for j in alpha_list :
        if i in j :
            dial_num = alpha_list.index(j) + 2
            num_list.append(dial_num)

print(sum(num_list)+len(num_list))
```
