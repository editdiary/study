# 1. [1330번] 두 수 비교하기

```python
A, B = map(int, input().split())

if A > B :
    print('>')
elif A < B :
    print('<')
else :
    print('==')
```

# 2. [9498번] 시험 성적

```python
grade = int(input())

if grade > 89 :
    print('A')
elif grade > 79 :
    print('B')
elif grade > 69 :
    print('C')
elif grade > 59 :
    print('D')
else :
    print('F')
```

# 3. [2753번] 윤년

```python
year = int(input())

if ((year % 4) == 0 and (year % 100) != 0) or (year % 400) == 0 :
    print(1)
else :
    print(0)
```

# 4. [14681번] 사분면 고르기

```python
x = int(input())
y = int(input())

if x > 0 :
    if y > 0 :
        print(1)
    else :
        print(4)
else :
    if y > 0 :
        print(2)
    else :
        print(3)
```

# 5. [2884번] 알람 시계

```python
H, M = map(int, input().split())

M -= 45

if M < 0 :
    H -= 1
    M += 60
    if H < 0 :
        H += 24

print(H, M)
```

# 6. [2525번] 오븐 시계

```python
A, B = map(int, input().split())
C = int(input())

B += C

while B >= 60 :
    B -= 60
    A += 1
    if A >= 24 :
        A -= 24

print(A, B)
```

# 7. [2480번] 주사위 세개

```python
a, b, c = map(int, input().split())

if a == b == c :
    prize = 10000 + a * 1000
elif a != b and a != c and b != c :
    prize = max(a, b, c) * 100
else :
    if a == b :
        prize = 1000 + a * 100
    else :
        prize = 1000 + c * 100

print(prize)
```
