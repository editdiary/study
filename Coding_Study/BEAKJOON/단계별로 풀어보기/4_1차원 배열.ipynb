# 5. [10810번] 공 넣기

```python
N, M = map(int, input().split())

num_list = [0] * N

for _ in range(M) :
    i, j, k = map(int, input().split())

    for n in range(i-1, j) :
        num_list[n] = k

print(*num_list)
```

# 6. [10813번] 공 바꾸기

```python
N, M = map(int, input().split())

num_list = [x for x in range(1, N+1)]

for _ in range(M) :
    i, j = map(int, input().split())
    num_list[i-1], num_list[j-1] = num_list[j-1], num_list[i-1]

print(*num_list)
```

# 9. [10811번] 바구니 뒤집기

```python
# My Solution
N, M = map(int, input().split())

num_list = [x for x in range(1, N+1)]

for _ in range(M) :
    i, j = map(int, input().split())

    rep = (j-i+1) // 2
    for n in range(rep) :
        num_list[i-1+n], num_list[j-1-n] = num_list[j-1-n], num_list[i-1+n]

print(*num_list)
```

```python
# Referenced Code
N, M = map(int, input().split())

num_list = [x for x in range(1, N+1)]

for _ in range(M) :
    i, j = map(int, input().split())
    num_list[i-1:j] = num_list[i-1:j][::-1]

print(*num_list)
```

# 10. [1546번] 평균

```python
# My Solution
N = int(input())

grade = list(map(int, input().split()))

M = max(grade)

for i in range(len(grade)) :
    grade[i] = grade[i]/M*100

print(sum(grade)/len(grade))
```

```python
# Referenced Code
N = int(input())

grade = list(map(int, input().split()))

M = max(grade)

grade = [x/M*100 for x in grade]

print(sum(grade)/len(grade))
```
