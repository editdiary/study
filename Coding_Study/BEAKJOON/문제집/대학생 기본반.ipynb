# [1747번] 소수&팰린드롬

```python
N = int(input())

def Palindrome(x) :
    x = str(x)
    if x == x[::-1] :
        return True
    return False

def prime_num(x) :
    if x == 1 :
        return False
    for i in range(2, int(x**(1/2)+1)) :
        if (x % i) == 0 :
            return False
    return True

while True :
    if Palindrome(N) and prime_num(N) :
        print(N)
        break
    else :
        N += 1
```

# [1652번] 누울 자리를 찾아라

```python
# My Solution
N = int(input())

room = [list(input()) for _ in range(N)]

place_x, place_y = 0, 0

for i in range(N) :
    key_x, key_y = False, False
    for j in range(N-1) :
        if (room[i][j] == '.') & (room[i][j+1] == '.') :
            key_x = True
        else :
            if key_x :
                place_x += 1
                key_x = False

        if (room[j][i] == '.') & (room[j+1][i] == '.') :
            key_y = True
        else :
            if key_y :
                place_y += 1
                key_y = False

    if key_x :
        place_x += 1
    if key_y :
        place_y += 1

print(place_x, place_y)
```

```python
# Referenced Code
N = int(input())

room = [input() for _ in range(N)]
room_h = [''.join(i) for i in zip(*room)]

place_x, place_y = 0, 0

for i in range(N) :

    for j in room[i].split('X') :
        if '..' in j :
            place_x += 1

    for k in room_h[i].split('X') :
        if '..' in k :
            place_y += 1

print(place_x, place_y)
```

# [2621번] 카드게임

```python
def cont_function(l) :
    l.sort()
    for i in range(len(l)-1) :
        if l[i] + 1 != l[i+1] :
            return False
    return True

def count_same_num(l) :
    temp_set = set(l)
    num_count = 0
    for i in temp_set :
        if num_count < l.count(i) :
            num_count = l.count(i)
            same_num = i
    return same_num, num_count

color = []
num = []

# 입력
for _ in range(5) :
    c, n = input().split()
    color.append(c)
    num.append(int(n))


if len(set(color)) == 1 :   # 색 동일한 경우

    continue_key = cont_function(num)      # 숫자 연속성 여부
    if continue_key :   # 색이 모두 같고, 숫자가 연속적 -> (1)
        score = max(num) + 900
    else :              # 색이 모두 같고, 숫자가 다름 -> (4)
        score = max(num) + 600

else :                      # 색 다른 경우

    continue_key = cont_function(num)      # 숫자 연속성 여부
    if continue_key :   # 색이 다르며, 숫자는 연속적 -> (5)
        score = max(num) + 500
    else :              # 색이 다르며, 숫자도 다름
        same_num, num_count = count_same_num(num)
        temp_list = [x for x in num if x != same_num]

        if num_count == 4 : # 색이 다르며, 4개의 숫자가 같은 경우 -> (2)
            score = same_num + 800

        elif num_count == 3 :
            if len(set(temp_list)) == 1 :   # 색이 다르며, 3개의 숫자가 같고, 나머지도 같은 경우 -> (3)
                score = same_num * 10 + temp_list[0] + 700
            else :  # 색이 다르며, 3개의 숫자가 같고, 나머지는 다른 경우 -> (6)
                score = same_num + 400

        elif num_count == 2 :
            same_num2, num_count2 = count_same_num(temp_list)
            if num_count2 == 2 :    # 색이 다르며, 2개의 숫자가 같고, 나머지 중 2개의 숫자가 같은 경우 -> (7)
                score = max(same_num, same_num2) * 10 + min(same_num, same_num2) + 300
            else :  # 색이 다르며, 2개의 숫자가 같고, 나머지는 다른 경우 -> (8)
                score = same_num + 200
        else :  # 전부 같지 않은 경우 -> (9)
            score = max(num) + 100

print(score)
```

# [1110번] 더하기 사이클

```python
N = int(input())

input_num = N
cycle = 0

while 1 :
    if N < 10 :
        N *= 11
    else :
        temp_num = N//10 + N%10
        N = int(str(N%10) + str(temp_num%10))
    cycle += 1

    if input_num == N :
        break

print(cycle)
```

# [2816번] 디지털 티비

```python
N = int(input())

chanel_list = [input() for _ in range(N)]
button_list = []

while 1 :
    if chanel_list[0] == 'KBS1' :
        if chanel_list[1] == 'KBS2' :
            break
        else :
            place_kbs2 = chanel_list.index('KBS2')
            button_list.append('1'* place_kbs2)
            button_list.append('4'* (place_kbs2 - 1))
            chanel_list.pop(place_kbs2)
            chanel_list.insert(1, 'KBS2')

    else:
        place_kbs1 = chanel_list.index('KBS1')
        button_list.append('1'* place_kbs1)
        button_list.append('4'* place_kbs1)
        chanel_list.pop(place_kbs1)
        chanel_list.insert(0, 'KBS1')

print(*button_list, sep='')
```

# [8979번] 올림픽

```python
# Referenced Code (https://develop247.tistory.com/234)
N, K = map(int, input().split())

medals = [list(map(int, input().split())) for _ in range(N)]

medals.sort(key = lambda x : (x[1], x[2], x[3]), reverse = True)

idx = [medals[i][0] for i in range(N)].index(K)

for i in range(N) :
    if medals[idx][1:] == medals[i][1:] :
        print(i+1)
        break
```
