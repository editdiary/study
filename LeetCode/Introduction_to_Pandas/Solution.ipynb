# Pandas Data Structures

## 2877. Create a DataFrame from List

```python
# My Solution
def createDataframe(student_data) -> pd.DataFrame:
    return pd.DataFrame(student_data, columns=['student_id', 'age'])

createDataframe(student_data)
```

# Data Inspection

## 2878. Get the Size of a DataFrame

```python
# My Solution
def getDataframeSize(players: pd.DataFrame):
    return list(players.shape)

getDataframeSize(players)
```

## 2879. Display the First Three Rows

```python
# My Solution
def selectFirstRows(employees: pd.DataFrame) -> pd.DataFrame:
    return employees.head(3)
```

# Data Selecting

## 2880. Select Data

```python
# My Solution
def selectData(students: pd.DataFrame) -> pd.DataFrame:
    return students[students['student_id'] == 101][['name', 'age']]
```

## 2881. Create a New Column

```python
# My Solution
def createBonusColumn(employees: pd.DataFrame) -> pd.DataFrame:
    employees['bonus'] = employees['salary'] * 2
    return employees
```

# Data Cleaning

## 2882. Drop Duplicate Rows

```python
# My Solution
def dropDuplicateEmails(customers: pd.DataFrame) -> pd.DataFrame:
    return customers.drop_duplicates('email')
```

## 2883. Drop Missing Data

```python
# My Solution
def dropMissingData(students: pd.DataFrame) -> pd.DataFrame:
    return students.dropna(subset = 'name')
```

## 2884. Modify Columns

```python
# My Solution
def modifySalaryColumn(employees: pd.DataFrame) -> pd.DataFrame:
    employees['salary'] = employees['salary'] * 2
    return employees
```

## 2885. Rename Columns

```python
# My Solution
def renameColumns(students: pd.DataFrame) -> pd.DataFrame:
    return students.rename(columns={'id': 'student_id', 'first': 'first_name', 'last': 'last_name', 'age': 'age_in_years'})
```

## 2886. Change Data Type

```python
# My Solution
def changeDatatype(students: pd.DataFrame) -> pd.DataFrame:
    students['grade'] = students['grade'].astype(int)
    return students
```

## 2887. Fill Missing Data

```python
# My Solution
def fillMissingValues(products: pd.DataFrame) -> pd.DataFrame:
    products['quantity'].fillna(0, inplace=True)
    return products
```

# Table Reshaping

## 2888. Reshape Data: Concatenate

```python
# My Solution
def concatenateTables(df1: pd.DataFrame, df2: pd.DataFrame) -> pd.DataFrame:
    return pd.concat([df1, df2], axis=0)
```

## 2889. Reshape Data: Pivot

```python
# My Solution
def pivotTable(weather: pd.DataFrame) -> pd.DataFrame:
    return pd.pivot_table(weather, index = 'month', columns='city', values='temperature', aggfunc = 'sum')
```

## 2890. Reshape Data: Melt

```python
# My Solution
def meltTable(report: pd.DataFrame) -> pd.DataFrame:
    return pd.melt(report, id_vars='product', var_name='quarter', value_name='sales')
```

# Advanced Techniques

## 2891. Method Chaining

```python
# My Solution
def findHeavyAnimals(animals: pd.DataFrame) -> pd.DataFrame:
    more_100 = animals[animals['weight'] > 100].sort_values('weight', ascending=False)
    return more_100[['name']]
```
