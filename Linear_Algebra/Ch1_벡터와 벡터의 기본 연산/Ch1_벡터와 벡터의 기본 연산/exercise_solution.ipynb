### Figure A

```python
# The vectors
v = np.array([1,2])
w = np.array([4,-6])
vPlusW = v + w

# now plot all three vectors
plt.figure(figsize=(8,8))

# plt.arrow(x, y, dx, dy, **kwargs): Add an arrow from (x, y) to (dx, dy) to the axes.
a1 = plt.arrow(0, 0, v[0], v[1], head_width=.3, width=.1, color='k', length_includes_head=True)                     # color = 'k' is black
a2 = plt.arrow(v[0], v[1], w[0], w[1], head_width=.3, width=.1, color=[.5, .5, .5], length_includes_head=True)      # color = [R, G, B]
a3 = plt.arrow(0, 0, vPlusW[0], vPlusW[1], head_width=.3, width=.1, color='tab:orange', length_includes_head=True)  # color = 'tab:orange' is orange in Tableau color palette

# make the plot look nice
plt.grid(linestyle='--', linewidth=.5)
plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.legend([a1, a2, a3], ['v', 'w', 'v+w'], fontsize=15)
plt.title('Vectors v, w, and v+w', fontsize=18)
plt.show()
```

### Figure B

```python
# vector difference
vMinusW = v - w

# now plot all three vectors
plt.figure(figsize=(8,8))

a1 = plt.arrow(0, 0, v[0], v[1], head_width=.3, width=.1, color='k', length_includes_head=True)
a2 = plt.arrow(0, 0, w[0], w[1], head_width=.3, width=.1, color=[.5, .5, .5], length_includes_head=True)
a3 = plt.arrow(w[0], w[1], vMinusW[0], vMinusW[1], head_width=.3, width=.1, color='tab:orange', length_includes_head=True)

# make the plot look nice
plt.grid(linestyle='--', linewidth=.5)
plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.legend([a1, a2, a3], ['v', 'w', 'v-w'], fontsize=15)
plt.title('Vectors v, w, and v-w', fontsize=18)
plt.show()
```

# Exercise 2

```python
# The function of calculating norm
def norm_of_vectors(v):
    return np.sqrt(np.sum(v**2))

# Test it on a unit-norm vector
w = np.array([0, 0, 1])
print(norm_of_vectors(w))
```

```python
# non-unit-norm vector, and confirm using np.linalg.norm
w = np.array([1, 2, 3])
print(norm_of_vectors(w), np.linalg.norm(w))
print(norm_of_vectors(w) == np.linalg.norm(w))
```

```python
# define function
def create_unit_vector(v):
    return v / norm_of_vectors(v)

# test on a unit vector
w = np.array([0, 1, 0])
print(create_unit_vector(w))

# test on a non-unit vector that is easy to confirm
w = np.array([0, 3, 0])
print(create_unit_vector(w))

# test on a non-unit vector
w = np.array([13, -5, 7])
unit_w = create_unit_vector(w)
print(unit_w)
# confirm the vectors' norms
print("Vector's norm: %f, Unit vector's norm: %f" %(np.linalg.norm(w), np.linalg.norm(unit_w)))

# what happens with the zeros vector?
print("\n")  # just some space
create_unit_vector(np.zeros((4,1)))
```

# Exercise 4

```python
# define the function
def create_mag_vector(v, mag):
    return mag * (v / np.linalg.norm(v))  # (v / np.linalg.norm(v)) is the unit vector in the direction of v

# test on a vector that is easy to confirm
w = np.array([1, 0, 0])
mag_vector = create_mag_vector(w, 4)
print(mag_vector)

# confirm the vectors' norms
print(f"norm of vector: {np.linalg.norm(w)}, norm of mag vector: {np.linalg.norm(mag_vector)}")
```

# Exercise 9

```python
# generate random R2 vectors (note: no orientation here! we don't need it for this exercise)
t = np.random.randn(2)
r = np.random.randn(2)

# the decomposition
t_para = r * (np.dot(t,r) / np.dot(r,r))  # parallel component
t_perp = t - t_para  # perpendicular component

# confirm that the two components sum to the target
print(t)
print(t_para + t_perp)

# confirm orthogonality (dot product must be zero)
print(np.dot(t_para, t_perp))
# Note about this result: Due to numerical precision errors, 
#   you might get a result of something like 10^-17, which can be interpretd as zero.
```

```python
# draw them!
plt.figure(figsize=(6,6))

# draw main vectors
plt.plot([0,t[0]],[0,t[1]],color='k',linewidth=3,label=r't')
plt.plot([0,r[0]],[0,r[1]],color='orange',linewidth=3,label=r'r')

# draw decomposed vector components
plt.plot([0,t_para[0]],[0,t_para[1]],'k--',linewidth=3,label=r't_para')
plt.plot([0,t_perp[0]],[0,t_perp[1]],'k:',linewidth=3,label=r't_perp')

plt.axis('equal')
plt.legend()
plt.show()
```
